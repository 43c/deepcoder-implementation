```python
from pprint import pprint

'''
@dataclasses.dataclass
class Entry:
    source_code: str
    examples: List[Example]
    attribute: Dict[str, bool]

@dataclasses.dataclass
class Example:
    """
    An I/O example

    Attributes
    ----------
    inputs : List[Primitive]
    output : Primitive
    """
    inputs: List[Primitive]
    output: Primitive

Primitive = Union[int, List[int]]
'''

print("SOURCE CODE")
print(dataset[0].source_code)
print()
print(f"num_examples={len(dataset[0].examples)}")
print()
print("first example")
pprint(dataset[0].examples[0], compact=True)
print()
```

```python
def line_count(p):
    lines = p.split("\n")
    count = 0
    for line in lines:
        if not (line.endswith("<- int") or line.endswith("<- [int]")):
            count += 1
    
    return count

line_count(dataset[0].source_code)
```

```python
line_counts = []

for entry in dataset:
    line_counts.append(line_count(entry.source_code))
```

```python
print(len(line_counts))


min_l = min(line_counts)
max_l = max(line_counts)

print(f"min_length = {min_l}")
print(f"max_length = {max_l}")
print()

count_for_length = []

for i in range(min_l, max_l + 1):
    tmp_count = sum([1 for count in line_counts if count == i])
    count_for_length.append(tmp_count)
    marker = f"i={i:>2} | "
    print(f"{marker}{tmp_count}")
```

```python
line_counts

file_name = "eval_set_line_counts.txt"

with open(file_name, "w") as file:
    for item in line_counts:
        file.write(f"{item},")
```

```python
index_lt3 = []

for i, c in enumerate(line_counts):
    if c <= 3:
        index_lt3.append(i)
```

```python
for k in range(1,9):
    tmp = []
    for i, c in enumerate(line_counts):
        if c <= k:
            tmp.append(str(i))

    to_write = ",".join(tmp)
    fn = f"eval_set_max_length_{k}.txt"
    with open(fn, "w") as file:
        file.write(f"{to_write}")
```
